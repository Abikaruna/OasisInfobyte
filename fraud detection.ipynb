```python
from imblearn.over_sampling import SMOTE
```

```python
X_res,y_res=SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Define classifiers
classifier = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree Classifier": DecisionTreeClassifier()
}

# Loop through each classifier, train and evaluate
for name, clf in classifier.items():
    print(f"\n********* {name} ***********")  # Print the name of the classifier
    clf.fit(X_train, y_train)  # Fit the classifier to the training data
    y_pred = clf.predict(X_test)  # Make predictions on the test set
    
    # Evaluate the classifier performance
    print(f"\nAccuracy: {accuracy_score(y_test, y_pred)}")
    print(f"\nPrecision: {precision_score(y_test, y_pred)}")
    print(f"\nRecall: {recall_score(y_test, y_pred)}")
    print(f"\nF1 Score: {f1_score(y_test, y_pred)}")
```
